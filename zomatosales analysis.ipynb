```python
df.drop_duplicates()
df.shape
```

# Cleaning rate column

```python
df['rate']
```

```python
# * IN the rate column there is 4.1/5 we donot require /5 *
```

```python
df['rate'].unique()
```

```python
# rate column has some values which are 'NEW' and 'nan' and '-'
#  we have to remove them also 
```

```python
def handlerate(value):
    if(value == 'NEW' or value =='-'):
        return np.nan  # converting the "NEW" and "-" values to nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate']
```

```python
df.rate.isnull().sum() # gives the number of null values in the chart
```

# Filling Null Values in Rate Column with Mean

```python
df['rate'].fillna(df['rate'].mean(),inplace = True) # filling the null values with mean values
df['rate'].isnull().sum()
```

```python
df.location.isnull().sum()
```

** Cusines location donot have so much null values so we can drop them**

# Dropping Null VALUES

```python
df.dropna(inplace = True)
df.head()
```

```python
df.location.isnull().sum()
```

# Renaming columns 

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates',
                    'listed_in(type)':'Type'},inplace = True)
df.head()
```

# Cleaning Rest Type Column

```python
rest_types = df['rest_type'].value_counts()
rest_types
```

there are some resturant types which have values very less so we are going to put them 
in the same resturant type called as others

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

# Making Rest Types less than 1000 in frequency

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df['rest_type'].unique()
```

# Cleaning location column where location < 300 making them others

```python
location = df['location'].value_counts(ascending = True)
location
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

# Cleaning cuisines column where cuisines ordred<100 are put into others

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

# Visualizing online order facility locationwise

By this we can see that which places ther online facility is count is more  and 
so that one can decide if a region has less online facility then we can open a resturant there 
where online facilty is not present

```python
locations_grouped_online = df.groupby(['location','online_order']).agg(cnt = ('online_order','count'))
locations_grouped_online = locations_grouped_online.reset_index()
locations_grouped_online
```

```python
locations_grouped_online = pd.pivot_table(locations_grouped_online,index=['location'],columns=['online_order'],
                                    fill_value=0)#aggfunc = np.sum, values=None,
locations_grouped_online = locations_grouped_online.reset_index()
locations_grouped_online
```

```python
import numpy as np
plt.figure(figsize= (11,4))
plt.xticks(rotation = 'vertical')
bar_width = 0.35
x=np.arange(len(locations_grouped_online['location']))
plt.bar(x+bar_width/2,locations_grouped_online['cnt']['No'],bar_width,color='orange')
plt.bar(x-bar_width/2,locations_grouped_online['cnt']['Yes'],bar_width,color='red')
plt.show()
```

```python
locations_grouped_online.plot(kind='bar',x='location',figsize=(11,4))
```
