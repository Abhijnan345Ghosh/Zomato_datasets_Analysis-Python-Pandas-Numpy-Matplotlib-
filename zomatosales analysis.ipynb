# Cleaning rate column

```python
df['rate']
```

```python
# * IN the rate column there is 4.1/5 we donot require /5 *
```

```python
df['rate'].unique()
```

```python
# rate column has some values which are 'NEW' and 'nan' and '-'
#  we have to remove them also 
```

```python
def handlerate(value):
    if(value == 'NEW' or value =='-'):
        return np.nan  # converting the "NEW" and "-" values to nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate']
```

```python
df.rate.isnull().sum() # gives the number of null values in the chart
```

# Filling Null Values in Rate Column with Mean

```python
df['rate'].fillna(df['rate'].mean(),inplace = True) # filling the null values with mean values
df['rate'].isnull().sum()
```

```python
df.location.isnull().sum()
```

** Cusines location donot have so much null values so we can drop them**

# Dropping Null VALUES

```python
df.dropna(inplace = True)
df.head()
```

```python
df.location.isnull().sum()
```

# Renaming columns 

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates',
                    'listed_in(type)':'Type'},inplace = True)
df.head()
```

# Listed in(city) and loaction both are same so dropping one column

```python
df = df.drop(['listed_in(city)'],axis = 1)
df.head()
```

```python
df['Cost2plates'].unique()
# this values are string having , but this must be float 
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

# Cleaning Rest Type Column

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts()
rest_types
```

```python
resttypegr1000 = rest_types[rest_types>1000]
resttypegr1000
```

```python
# there are some resturant types which have values very less so we are going to put them 
# in the same resturant type called as others
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

# Making Rest Types less than 1000 in frequency

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df['rest_type'].unique()
```

# Cleaning location column where location < 300 making them others

```python
location = df['location'].value_counts(ascending = True)
location
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

# Cleaning cuisines column where cuisines ordred<100 are put into
# others

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
